## Q8.Now suppose I have a budget of 200k and I can buy more than one model of phones , instead of just one model . What do I do ? Do I still buy the same model's phones or different phones ? Give me the ideal distribution of models of phones I should buy for maximum sales .

```python
ans6_df
```

To buy the maximum mobile with limited buget the list of mobile should be according to price first

```python
ans6_df = ans6_df.sort_values(by='price', ascending=True)
```

```python
ans6_df
```

```python
order_list = []
amount = 200000
min_amt=ans6_df['price'].min()
while(amount >= min_amt):
    for index, row in ans6_df.iterrows():
        price = row['price']

        if amount != 0 and price <= amount:
            order_list.append(row['title'])
            amount -= price
        else:
            break

order_list
```

```python
amount
```

```python
# Convert to a set to get unique values
unique_values = set(order_list)

# Convert back to a list 
unique_order_list = list(unique_values)
```

```python
unique_order_list
```

```python
from collections import Counter
counter=Counter(order_list)
```

```python
counter
```

```python
ans8_df = pd.DataFrame(list(counter.items()), columns=['Title', 'No._of_Products'])
```

```python
ans8_df
```
